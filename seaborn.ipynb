```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

http://www.wawwj.com/2014/_SP/ran_wine.php

```python
wty = pd.read_csv('wawwj2014.csv')
wty = wty.rename(columns={'País':'Country',
                          'Bodega':'Brand',
                          'Vino':'Wine',
                          'Tipo':'Type',
                          'Premios':'Prizes',
                          'Ptos':'Points',
                          '#': 'ID'})
wty = wty.set_index('ID')
wty.Country = wty.Country.map(str.upper)
wty.Type = wty.Type.map(str.upper)
wty.Brand = wty.Brand.map(str.upper)
```

```python
sb.set_context("talk")

sb.set(style="ticks")

wty_country_prizes = wty['Prizes'].groupby(wty['Country'])
idx = wty_country_prizes.count() == 1
wty_t = wty.ix[[c not in idx[idx].index for c in wty.Country], ['Country', 'Prizes']]

wty_t_country_mean = wty_t.groupby(wty_t.Country).mean()
country_names = wty_t_country_mean.Prizes.sort(inplace=False).index

f, ax = plt.subplots()
sb.offset_spines()
sb.violinplot(wty_t['Prizes'], wty_t['Country'], order=country_names)
locs, labels = plt.xticks()
plt.setp(labels, rotation=45)
sb.despine(trim=True)
```

```python
f, ax = plt.subplots()
sb.offset_spines()
sb.boxplot(wty_t['Prizes'], wty_t['Country'], order=country_names)
locs, labels = plt.xticks()
plt.setp(labels, rotation=45)
sb.despine(trim=True)
```
