```python
%matplotlib inline
import pandas as pd
import numpy as np
from ggplot import *
```

```python
wty = pd.read_csv('wawwj2014.csv')
wty = wty.rename(columns={'País':'Country',
                          'Bodega':'Brand',
                          'Vino':'Wine',
                          'Tipo':'Type',
                          'Premios':'Prizes',
                          'Ptos':'Points',
                          '#': 'ID'})
wty = wty.set_index('ID')
wty.Country = wty.Country.map(str.upper)
wty.Type = wty.Type.map(str.upper)
wty.Brand = wty.Brand.map(str.upper)
```

```python
ggplot(wty, aes(x='Prizes', y='Points')) + geom_point(size=60)
```

```python
ggplot(wty, aes(y="Points", x="Prizes")) + \
    geom_point(size=60, shape=6, alpha=0.4) + \
    stat_smooth(method='lm', colour='red', size=4, se=False)
```

```python
ggplot(wty, aes("Country")) + geom_bar() + theme(axis_text_x=element_text(angle=90))
```

```python
ggplot(wty, aes("Country")) + \
    geom_bar(fill="red", colour="black") + \
    theme(axis_text_x=element_text(angle=90))
```

```python
ggplot(wty, aes("Country")) + \
    geom_bar(fill="red", colour="black") + \
    theme_xkcd() + \
    theme(axis_text_x=element_text(angle=90))
```

```python
ggplot(wty, aes("factor(Prizes)")) + \
    geom_bar() + \
    facet_wrap('Country', scales='fixed')
```

```python
ggplot(wty, aes(y="Points", x="Prizes")) + \
    geom_point() + \
    facet_wrap("Country", scales="fixed")
```

```python
idx = (wty.Country == 'AUSTRALIA') | \
    (wty.Country == 'FRANCE') | \
    (wty.Country == 'SPAIN') | \
    (wty.Country == 'USA') | \
    (wty.Country == 'ARGENTINA')

ggplot(wty[idx], aes(x="Prizes", y="Points", colour="Country")) + \
    geom_point() + \
    stat_smooth(method='lm', color='blue') + \
    facet_wrap("Country", scales="fixed")
```
